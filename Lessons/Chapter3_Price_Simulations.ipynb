```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 1000
P0 = 10
mu = 0.001
```

```python
last_p = np.log(P0)
```

```python
log_returns = np.zeros(T)
prices = np.zeros(T)
```

```python
for t in range(T):
    r = 0.01 * np.random.randn()
    p = last_p + mu + r
    log_returns[t] = r + mu
    prices[t] = np.exp(p)
    last_p = p
```

```python
plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title('Simulated prices')
```
